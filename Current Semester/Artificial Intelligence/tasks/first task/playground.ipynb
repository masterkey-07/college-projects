```python
import pandas as pd
import math as m
import matplotlib.pyplot as plt
import numpy as np
```

```python
train = pd.read_csv('./data/train.csv')

new_result = pd.read_csv('./result/new_result.csv', sep=';')  

new_result_length = len(new_result.index)

the_best = new_result[new_result['accuracy'] > 0.86]

for c in the_best.columns.values.tolist():
    print(c)
```

```python
def plot_graph_bar(column:str):
    column_data = the_best[column].value_counts()
    # uniques = sorted(column_data.unique())
    # Criando o gráfico de barras para max_depth
    plt.figure(figsize=(12, 6))  # Tamanho da figura
    # plt.hist(column_data, color='lightgreen', edgecolor='black', bins=uniques)  # Criando o histograma
    column_data.plot(kind='bar', color='lightgreen', edgecolor='black')  # Criando o gráfico de barras
    plt.title(f'Distribuição de {column}')  # Título do gráfico
    plt.xlabel(column)  # Rótulo do eixo x
    plt.ylabel('Frequência')  # Rótulo do eixo y

    # Exibindo os gráficos
    plt.tight_layout()  # Ajusta o layout para evitar cortar as legendas
    plt.show()
```

```python
plot_graph_bar('splitter')
```

```python
plot_graph_bar('criterion')
```

```python
# Sample data
categories = ['Não Sobreviveram', 'Sobreviveram']

not_survived = train[train['Survived'] == 0]
survived = train[train['Survived'] == 1]

male_not_survide = not_survived[not_survived['Sex'] == 'male']['Sex'].count()
female_not_survide = not_survived[not_survived['Sex'] == 'female']['Sex'].count()

male_survide = survived[survived['Sex'] == 'male']['Sex'].count()
female_survide = survived[survived['Sex'] == 'female']['Sex'].count()

values1 = [male_not_survide, female_not_survide]
values2 = [male_survide, female_survide]

print(values1[0]/ (values1[0] + values2[0]))
print(values1[1]/ (values1[1] + values2[1]))

print(values1[1]/ (values1[0] + values1[1]))
print(values2[1]/ (values2[0] + values2[1]))

# Width of the bars
bar_width = 0.35

# Creating the bar graph
plt.figure(figsize=(10, 6))
bar1 = plt.bar([i - bar_width/2 for i in range(len(categories))], values1, bar_width, color='skyblue', label='Homens')
bar2 = plt.bar([i + bar_width/2 for i in range(len(categories))], values2, bar_width, color='orange', label='Mulheres')

# Adding titles and labels
plt.title('Gráfico dos Sobreviventes dividido por Sexo')
plt.ylabel('Quantidade')
plt.xticks(range(len(categories)), categories)
plt.legend()

# Displaying the plot
plt.tight_layout()
plt.show()
```
