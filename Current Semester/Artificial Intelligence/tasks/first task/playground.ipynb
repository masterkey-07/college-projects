```python
def plot_graph_bar(column:str):
    column_data = the_best[column].value_counts()
    # uniques = sorted(column_data.unique())
    # Criando o gráfico de barras para max_depth
    plt.figure(figsize=(12, 6))  # Tamanho da figura
    # plt.hist(column_data, color='lightgreen', edgecolor='black', bins=uniques)  # Criando o histograma
    column_data.plot(kind='bar', color='lightgreen', edgecolor='black')  # Criando o gráfico de barras
    plt.title(f'Distribuição de {column}')  # Título do gráfico
    plt.xlabel(column)  # Rótulo do eixo x
    plt.ylabel('Frequência')  # Rótulo do eixo y

    # Exibindo os gráficos
    plt.tight_layout()  # Ajusta o layout para evitar cortar as legendas
    plt.show()
```

```python
plot_graph_bar('splitter')
```

```python
plot_graph_bar('criterion')
```

```python
plot_graph_bar('max_depth')
```

```python
plot_graph_bar('min_sample_split')
```

```python
# Sample data
categories = ['Category 1', 'Category 2', 'Category 3', 'Category 4', 'Category 5']
values1 = [20, 35, 30, 25, 40]
values2 = [25, 30, 35, 20, 45]

# Width of the bars
bar_width = 0.35

# Creating the bar graph
plt.figure(figsize=(10, 6))
bar1 = plt.bar([i - bar_width/2 for i in range(len(categories))], values1, bar_width, color='skyblue', label='Bar 1')
bar2 = plt.bar([i + bar_width/2 for i in range(len(categories))], values2, bar_width, color='orange', label='Bar 2')

# Adding titles and labels
plt.title('Bar Graph Example')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.xticks(range(len(categories)), categories)
plt.legend()

# Displaying the plot
plt.tight_layout()
plt.show()
```

```python
from scipy import stats
import numpy as np

# Generate example data
binary_variable = np.random.choice([0, 1], size=100)  # Example binary variable
continuous_variable = np.random.randn(100)  # Example continuous variable

# Calculate Point-Biserial Correlation
point_biserial_corr, p_value = stats.pointbiserialr(binary_variable, continuous_variable)

print("Point-Biserial Correlation coefficient:", point_biserial_corr)
print("P-value:", p_value)
```
