```python
import numpy as np

# data processing
import pandas as pd

# Algorithms
from sklearn.tree import DecisionTreeClassifier
```

```python
from google.colab import drive
drive.mount('/content/drive')

from zipfile import ZipFile
zip_file = ZipFile('/content/drive/MyDrive/Disciplinas/2024/IA1s2024/Data/titanic.zip')

data_train = pd.read_csv(zip_file.open('train.csv'))
data_test = pd.read_csv(zip_file.open('test.csv'))
data_gender = pd.read_csv(zip_file.open('gender_submission.csv'))

alldata = [data_train, data_test]
```

```python
common_value = 'S'

for dataset in alldata:
    dataset['Embarked'] = dataset['Embarked'].fillna(common_value)
```

```python
data_train['Embarked'].isna().sum()
```

```python
for dataset in alldata:
    dataset['Fare'] = dataset['Fare'].fillna(0)
    dataset['Fare'] = dataset['Fare'].astype(int)
```

```python
genders = {'male': 0, 'female': 1}

for dataset in alldata:
    dataset['Sex'] = dataset['Sex'].map(genders)
```

```python
ports = {"S": 0, "C": 1, "Q": 2}

for dataset in alldata:
    dataset['Embarked'] = dataset['Embarked'].map(ports)
```

```python
data_train = data_train.drop(['PassengerId','Name','Ticket','Cabin', 'Age', 'SibSp'], axis=1)
data_test  = data_test.drop(['PassengerId','Name','Ticket','Cabin','SibSp', 'Age'], axis=1)
```

```python
from sklearn.tree import DecisionTreeClassifier

clf = DecisionTreeClassifier(random_state=0, max_depth=1)

print(data_train.drop(['Survived'], axis=1).columns.values)

X = data_train.drop(['Survived'], axis=1).values
Y = data_train['Survived'].values

clf.fit(X,Y)
```

```python
Y_test = data_gender['Survived'].values
X_test = data_test.values

print(data_test.columns.values)

Y_pred = clf.predict(X_test)
```

```python
from sklearn.metrics import balanced_accuracy_score
balanced_accuracy_score(Y_test, Y_pred)
```
